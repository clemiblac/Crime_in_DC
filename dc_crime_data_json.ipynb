```python
import pandas as pd
import json
from sqlalchemy import create_engine
import pymysql
from config import (username,password, host, database)
```

```python
target = f"mysql://{username}:{password}@{host}/{database}"
engine = create_engine(target,echo = True)
```

```python
connection = engine.connect()
for i in range(2010, 2020, 1):
    year = str(i)
    if year == "2010":
        dc_crime = pd.read_sql("SELECT * FROM crimeDB.crime2010", connection)
    else:
        crime_year = pd.read_sql(f"SELECT * FROM crimeDB.crime{year}", connection)
        dc_crime = dc_crime.append(crime_year, ignore_index = True)
```
