```python
dates_2020=[]
```

```python
dates = range(1, 32)
months = range(1, 9)
year = 2020

for i in months:
    for j in dates:
        date_list='{}-{}-{}'.format(year, i, j)
        #print(date_list)
        dates_2020.append(date_list)
        
```

```python
for i in dates_2020:
    #add leading zero to months
    mon=i[:5]+'0'+i[5:]
    print(mon)
    #add leadind zeros to dates less than 10
    
```

```python
#List of dates from beginning of the year to today
dates_2020
```
